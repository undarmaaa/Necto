# Clear Event Histograms
This notebook loads labeled clear events and visualizes histograms of key metrics (e.g., Î”ScoreProb) split by label (good/bad).

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Path to your labeled clears file
jsonl_path = "rlbot-support/Nexto/nexto_clears_labeled.jsonl"

# Load all clears into a DataFrame
records = []
with open(jsonl_path, "r", encoding="utf-8") as f:
    for line in f:
        try:
            record = json.loads(line)
            records.append(record)
        except Exception as e:
            print("Skipping line due to error:", e)

df = pd.json_normalize(records)

# Show available columns
print("Columns:", df.columns.tolist())
```

```python
# Example: Plot histogram of Î”ScoreProb (replace with actual column name if different)
scoreprob_col = None
for col in df.columns:
    if "score" in col.lower() or "delta" in col.lower():
        scoreprob_col = col
        break

if scoreprob_col:
    plt.figure(figsize=(8, 5))
    sns.histplot(data=df, x=scoreprob_col, hue="label", bins=20, kde=True, stat="density", common_norm=False)
    plt.title("Histogram of score by Label")
    plt.show()
else:
    print("Could not find Î”ScoreProb column. Please check the column names above.")
```

```python
# You can also plot other features, e.g. ball position, velocity, boost totals, etc.
# Example for ball.x at clear event (if available):
if "clear_event.ball.x" in df.columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(data=df, x="clear_event.ball.x", hue="label", bins=20, kde=True, stat="density", common_norm=False)
    plt.title("Histogram of Ball X Position at Clear by Label")
    plt.show()
```
